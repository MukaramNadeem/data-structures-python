```python
#Factorial
def fact(n):
    #base case
    if n<= 1:
        return 1

    #induction case
    return n * fact (n-1) #tail recursion : calculation in the return statement that is(*)
```

```python
for i in range(1,7):
    print("Factorial of : " , i , "is",fact(i))
```

```python
#Fibonaci sequence algorithm using recusion 

def fib(n):
    if n<=1:
        return 1

    return fib(n-1) + fib (n-2) 
```

```python
%timeit fib(10)
```

```python
#Fibonaci sequence algorithm using iteration 
def fib(n):
    a,b = 1,1
    for i in range (n):
        a,b = b , a+b
    return a
```

```python
fib(20)
```

```python
#Fibonaci sequence algorithm using recursion but fast : its essencially same as iteration because its iterating the function call.
def fib(n,a=1,b=1):
    if n ==0:
        return a

    return fib(n-1,b,a+b) #not tail recursion : No operation in the return statement
```

```python
fib(20)
```

```python
def towerOfHanoi(level = 3):
    moveTower(level ,"A","C","B") # move the tower from A to C using B as auxilary

def moveTower(l,fr,to,ax):
    if l == 1:
        printMove(l,fr,to)
        return

    moveTower(l-1,fr,ax,to)
    printMove(l,fr,to)
    moveTower(l-1,ax,to,fr)

def printMove(l,fr,to):
    print("Move:",l,"from",fr,"to",to)
   
```

```python
def sum_list(l):
    sum = 0
    for i in l:
        sum += i
    return sum
```

```python
l = [1,3,4,5,6,2,8]
```

```python
sum_list(l)
```

```python
#using recursion : functioal programming

def sum_list_recursion(l):
    if len(l) == 0:
        return 0

    return l[0] + sum_list_recursion(l[1:]) #sending head + tail of list ###(l[1:]) this returns list except 1 element

```

```python
sum_list_recursion(l)
```
