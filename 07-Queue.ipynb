```python
class Queue:
    
    def __init__(self):
        self.size = 5
        self.q = list(range(self.size)) #python list of length 5 with some dummy values : "range"!!!

        self.i = 0 #to keep track of input  : #initilized at zero index
        self.o = 0 #to keep track of output  : #initilized at zero index
        #we use them so we can compare them for "is_full" and "is_empty" values

        self.is_empty = True #at first queue is empty
        self.is_full = False #at first queue is not full
```

```python
def _inc (self,index): #private method!!!
    #when val is at last index.
    if index + 1 == self.size: #first check if size is full (as we know self.size==5) then return to the "first index" of queue!!!
        return 0 #first index of queue
   
    
    else:   #otherwise add one in the value to move to next index
        return index + 1

Queue._inc = _inc
```

```python
def enqueue(self,val):
    #first check!
    if self.is_full:
        raise IndexError("Queue is full !!!")
    
    #now as the queue is not full, SO add value at i and increment it to move i to next index.
    self.q[self.i] = val
    self.i = self._inc(self.i)

    
    #last check!!! 
    if self.i == self.o:
        self.is_empty = False

Queue.enqueue = enqueue
```

```python
def dequeue(self):
    #first check!!!
    if self.is_empty:
        raise IndexError ("Queue is empty !!!")
    #return (donot remove) one value from o index and increment (move it to next index) it.
    ret = self.q[self.o]
    self.o = self._inc(self.o)

    #if it came back to i after returning values in circular way then the queue became empty.
    if self.i == self.o:
         self.is_empty = True
    #as we removed value so now its not full
    
    self .is_full = False
    
    
    return ret

Queue.dequeue = dequeue
```

```python
def __str__(self):
    return str (self.q) + ", i= "+ str(self.i) +", o= " +str(self.o)

Queue.__str__ = __str__
```
