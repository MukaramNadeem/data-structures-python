```python
def bubbleSort(l):
    #Compare first two elements and move the bigger element to right then compare the next two elements and move the bigger to right side and so on for the entire list.
    #Every time the loop ends the largest one moves to the end.
    #Thus the smaller element comes out to the top or start of the list like bubbles in water.
    
    n = len(l)

    for i in range(n):
        
        #swapped = False                         #An optional flag to end the loop earliar when the elements get sorted before the total number of defined loops.
        
        for j in range(0,(n-i)-1):               #If we donot use (-1) then IndexError (Index out of Bond)  for last element will occur.# (-i) so it donot compare the already sorted(biggest at end) ones.

            if l[j] > l[j+1]:
                l[j] , l[j+1] =  l[j+1] , l[j]
                #swapped = True

        #if not swapped:
            #break
```

```python
l = [1,2,3,4,5,6,1,110,15]
bubbleSort(l)
print(l)
```

```python
def insertionSort(l):
    #take the 0th index element out by using a variable "KEY"...
    #We will insert the key in the already sorted list at the left(already sorted list is always in left) at the correct place.Correct place will be found by looping through the already sorted values at left. 

    
    for i in range (1,len(l)):                 #Start the loop from 1st index till to the end of the list. 

        key = l[i]

        j = i-1                              #starting j one less than i.
        while j >= 0 and key < l[j]:         #to handle (-1) index case we used ` j >= 0 ` : As soon as it reach -1 the loop exit.
            l[j+1] = l[j]
            j-=1

        #when the loop end...l[j+1] is always empty
        l[j+1] = key
```

```python
l = [1,2,3,4,5,6,1,110,15]
insertionSort(l)
print(l)
```

```python
def selectionSort(l):
    n = len(l)
    
    #for each element in list (starting from left)
    for i in range(n):
        min_idx = i      #find the minimum
                         #in the "rest" of the list 
        for j in range(i+1,n):
            if l[j] < l[min_idx]: #change ` < ` for decending order
                min_idx = j
        #swap the minimum with current element,now we have (sorted stuff till i).
        l[i] , l[min_idx] = l[min_idx] , l[i]
```

```python
l = [1,2,3,4,5,6,1,110,15]
selectionSort(l)
print(l)
```

```python
import random

def quickSort(l,fst,lst):  
    """First we select a pivot randomly from whole list than We divide the whole list into two partitions one partition  
       has all elements smaller than pivot and the other has all elements greater than pivot using i and j than we pass
       these partitions in a recursive call and this process goes on for that partitions until the base case is reached"""
    
    if fst >= lst: #base case
        return 

    i , j = fst,lst #because fst,lst  will enentually divide(change) so we store them in i,j and change i,j and not the fst,lst. 

    pivot = l[random.randint(fst,lst)] #we can also pick first element as pivot : ` pivot = l[fst]`

    while i <= j:
        while l[i] < pivot:
            i+=1
        while l[j] > pivot:
            j-=1
        if i<=j:
            l[i],l[j] = l[j],l[i] #swapping values when both i,j reached
            i , j = i+1 , j-1     #after swapping we move them one more step in the same way. 

    quickSort(l,fst,j)
    quickSort(l,i,lst)
```

```python
l = [1,2,3,4,5,6,1,110,15]
quickSort(l,0,len(l)-1)
print(l)
```

## For Decending sort

```python
def selectionSort(l):  #Bad idea : Do not change the already implemented sort algorithm of a language. Its against abstraction as well.
    n = len(l)
    
    #for each element in list (starting from left)
    for i in range(n):
        min_idx = i      #find the minimum
                         #in the "rest" of the list 
        for j in range(i+1,n):
            if l[j] > l[min_idx]:  #!!! changed ` < ` to ` > ` for decending sort
                min_idx = j
        #swap the minimum with current element,now we have (sorted stuff till i).
        l[i] , l[min_idx] = l[min_idx] , l[i]
```

```python
l = [1,53,2,56,77,53,3,0]
selectionSort(l)
print(l)
```

## Better Way

```python
def less_than(a,b): #predicate function #its for accending sort
    return a<b
```

```python
def selectionSort(l,compare_with): #will pass a function in place of compare_with.
    n = len(l)

    for i in range(n):
        min_idx = i 

        for j in range(i+1,n):
            if compare_with(l[j] ,l[min_idx]): #instead of `direct comparison` we passed a function that will do comparison 
                min_idx = j

        l[i] , l[min_idx] = l[min_idx] , l[i]
```

```python
l = [1,3,4,5,6,44,22,55]
selectionSort(l,less_than) #do not use `less_than()` !!! we are not calling it!!! 
print(l)
```

```python
def greater_than(a,b):
    return a > b
```

```python
selectionSort(l,greater_than)
print(l)
```

## Taking it Even Further

```python
all_tupules = [(24,25), #made a list of tupules
    (1,2),
    (2,4),
    (3,5),
    ]
```

```python
def tupule_less_than(a,b):
    return sum(a) < sum (b) 
def tupule_greater_than(a,b):
    return sum(a) > sum(b)
    
```

```python
print("Ascending : \t ", end = " ")
selectionSort(all_tupules,tupule_less_than)
print(all_tupules)


print("Decending : \t ", end = " ")
selectionSort(all_tupules,tupule_greater_than)
print(all_tupules)
```

## Sorting in Python

## Sorting Objects of Custom classes and Lambda functions

```python
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __str__(self):
        return self.name + " : " + str(self.age)
```

```python
s1 = Student("Wajid",5)
s2 = Student("Majid",4)
s3 = Student("Sajid",9)
s = [s1,s2,s3] 
```

```python
for i in s:
    print(i)
```

```python
#Lambda functions are anonymous functions they donot have any name : So we do not need to think about their names. 
s.sort(key = lambda x : x.age)
```

```python
for i in s:
    print(i)
```

```python
s.sort(key = lambda x : x.age , reverse = True) #For decending
for i in s:
    print(i)
```
