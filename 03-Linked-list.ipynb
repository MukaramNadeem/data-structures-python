```python
class Node:   #node has only two things
    def __init__(self,val = None ):     
        self.val  = val
        self.next = None    #this "next" holds the address of next node

class LinkedList:    #a linked list is just a varaiable(head) that holds address of nodes
    def __init__(self):
        self.head = None
```

```python
def push(self,val): #push returns nothing
    new_node = Node(val) 
    if self.head is None: #case (i)
        self.head = new_node
        return

    #otherwise case (ii)
 
    last = self.head
    while last.next is not None:
        last = last.next

    
    last.next = new_node



LinkedList.push = push
```

```python
def pop(self): #pop returns a value
    #(i)
    if self.head is None:
        raise Exception ("No value in the list.")
    
    #(ii)
    if self.head.next is None:
        val = self.head.val
        self.head = None     #make the list empty
        return val
    
    #(iii) when there are more than one nodes. Use "two" pointers: make "prev" = temp and "temp" = next
    temp = self.head
    while temp.next is not None:
        prev = temp
        temp = temp.next
    #return the last value and make its next None using prev.next
    val = temp.val
    prev.next = None
    return val


LinkedList.pop = pop
```

```python
def insert(self,index,val):
    new_node =  Node(val)

    #(i)
    if index == 0:
        new_node.next = self.head #do this first otherwise you will loss the head (whole Linked list). 
        self.head = new_node
        return

    #(ii)
    temp = self.head

    counter = 0
    while temp is not None and counter < index: #focus on "and" : Loop will run till the given index tion and the end of the List.First check "None" condition to deal with this condion insert(10000,1).
        prev = temp
        temp = temp.next
        counter +=1
        
    prev.next = new_node
    new_node.next = temp
    
    #(iii) It will be aumatically handled by case(ii) logic because of "None".
        
LinkedList.insert = insert 
```

```python
def remove(self,val):
    
    temp = self.head

    #(i)
    if temp is not None:
        if temp.val == val:
            self.head = temp.next
            temp = None #not needed
            return
    #(ii)
    while temp is not None:
        if temp.val == val:
            break #value found

        prev = temp
        temp = temp.next

    if temp is None: #not found after loop execution
            return

    
    prev.next = temp.next  #or prev.next = prev.next.next


LinkedList.remove = remove
    
```

```python
def printList(self):
    ret_str = "["
    
    temp = self.head
    while temp is not None:
        ret_str += str(temp.val) + ","
        temp = temp.next
    
    ret_str = ret_str.rstrip(', ')
    ret_str += "]"
    return ret_str
LinkedList.__str__ = printList
```
