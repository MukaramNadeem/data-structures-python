## Maps HashMaps Dictionaries

```python
class HashMap: 
    def __init__(self):
        self.size = 10
        self.map = [None] * self.size #create a list of size(10) with all values initilized by None.It will then hold addresses of other *lists* 
```

```python
def _get_hash(self,key):
    return key % self.size        #will return value from 0-9 as  size is 10

HashMap._get_hash = _get_hash
```

```python
def add(self,key,value):
    key_hash = self._get_hash(key)
    key_value = [key,value] #key_value is a list that will hold a key(at index 0) and a value (at index 1).
    

    #if nothing present(means its a new hash) simply insert the *list* [key_value].
    if self.map[key_hash] is None:
        self.map[key_hash] = [key_value] #we are inserting a list into another list!!!
        return True

    
    #else case.
    for pair in self.map[key_hash]:
        if pair[0] == key:
            print("updating : ," , key)
            pair[1] = value
            return True
            
    #nope its a collision : insert it
    self.map[key_hash].append(key_value)
    return True

HashMap.add = add
```

```python
def get(self,key):
    
    key_hash = self._get_hash(key)       #so there is no need to iterate to reach key index.
    
    if self.map[key_hash] is not None:
        for pair in self.map[key_hash]:
            if pair[0] == key:          
                return pair[1]           #we  return only value.
    #if not found after iteration
    raise KeyError(str(key))

HashMap.get = get
```

```python
def delete(self,key):
    key_hash = self._get_hash(key)

    if self.map[key_hash] is None:
        raise KeyError(str(key))

    for i in range(0,len(self.map[key_hash])):  #its the way to get length of a list inside a list  ` self.map[key_hash] `
        if self.map[key_hash][i][0] == key:     #ith list ka 0th index
            self.map[key_hash].pop(i)           #pop(i) removes value at ith index.Here i is that whole key_value list
            return True

HashMap.delete = delete
        
```

```python
def __str__(self):
    ret = ""
    for i , item in enumerate(self.map):
        if item is not None:
            ret += str(i) + " : " + str(item) + "\n"
    return ret

HashMap.__str__ =__str__
```

```python
h = HashMap()
```

```python
h.add(17,"seventeen")
h.add(26,"sixteen")
h.add(35,"thirty five")
h.add(26,"twenty six updated")
h.add(36,"new value at same hash")
```
