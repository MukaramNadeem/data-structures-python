```python
class Stack:      #stack made by a python list with certain limitations.
    
    
    def __init__(self): #making a python list in constructor
        self.l = []   

    def push(self,val):
        self.l.append(val)

    def pop(self):
        return self.l.pop()
    
    def peek(self): #in a stack we can peek last value : it will not be removed.
        return self.l[-1] 
```

```python
s = Stack()
s.push(1)
s.push(2)
s.push(3)

#print(s[1]) #now its an error (good for us : as its a stack)
print(s)
print(s.pop())


#print(s.l[1]) we can still do this. But it will break our stack concept(abstraction).
```

```python
print(s.peek())
print(s.pop())
print(s.peek())
print(s.pop())
```

```python
def is_matched(string):
    opening = "({["
    closing = ")}]"
    mapping = dict (zip (opening,closing)) 

    
    stack = []
    
    
   
    for c in string:    #loop through whole string
        #Case 1:
        if c not in mapping.values() and c not in mapping.keys():
            continue
        
        #Case 2: This will(automatically) check only starting Brackets
        if c in mapping:
            stack.append(mapping[c]) #store them in stack to compare them with list
         
        
        #Case 3: Has the closing brackets if we reached till here
        elif len(stack) == 0 or c!= stack.pop(): # "  len(stack)==0  " this check is for edge cases("Extra brackets at the End") when stack becomes empty and stack.pop() gives error
            return False


    #Check if the stack is empty : All values got checked 
    #Only then return True : is_matched.
    return len(stack) ==0                         #===>    #      if len(stack) == 0:
                                                           #         return True
                                                           #      else:
                                                           #         return False
```

```python
str = "{2+(3*5) * (2*2) +5 }"
is_matched(str)
```

```python
str = "{2+(3*5) {* ((2*2) +5 )}"
is_matched(str)
```
