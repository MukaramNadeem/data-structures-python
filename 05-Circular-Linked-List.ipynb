```python
class Node:
    def __init__(self , data = None):
        self.val = data
        self.next = None


class Ring:
    def __init__(self):
        self.head = None
        
```

```python
def printList(self):
    ret_str = "["
    temp = self.head
    while temp is not None:
        ret_str += str(temp.val) + ", "
        temp = temp.next

        if temp == self.head: #change!!! 
            break

    ret_str = ret_str.rstrip(", ")
    ret_str += "]"
    return ret_str

Ring.__str__ = printList
```

```python
def _get_last(self): #a function to get last value of list: 
    #if there is no node : no last 
    if self.head is None:
        return None
        
    #if only one node: its last as well
    if self.head.next == self.head:
        return self.head

    
    #if more than one node
    temp = self.head.next # loop from 2nd node!!!
    while temp.next != self.head:
        temp = temp.next

    return temp

Ring._get_last = _get_last
```

```python
def insert(self,index,value):
    
    new_node = Node(value)
    
    last = self._get_last()   #change!!!

    #insert at index 0 is different
    if index == 0:

        new_node.next = self.head
        self.head = new_node

        #also set the last pointer to this new head #change!!!
        if last is None:
            self.head.next = self.head #first node being inserted

        else:
            last.next = new_node
    
        return


    #same error was present in other lists but not handeled.
    if self.head is None: #and index >0:
        raise IndexError("Cannot insert at " + str(index) + ": List is empty ")


    
    #for other indices
    #essencially same no change : value will be inserted after looping (circular or ring behaiour )
    temp = self.head 

    counter = 0

    while temp is not None and counter < index:
        prev = temp
        temp = temp.next
        counter += 1

    prev.next = new_node
    new_node.next = temp

Ring.insert = insert
```

```python
r = Ring()
#r.insert(0,1)
r.insert(0,1)
r.insert(0,2)
r.insert(1,3)
r.insert(7,5)
print(r)
```

```python
def remove(self , val):
    temp = self.head
    last = self._get_last() #change!!!
    #(i)
    
    #first node matches case : value fount at first index
    if temp.val == val:
        if last == self.head:
            self.head = None
        else:
            self.head = temp.next
            last.next = self.head
        return


        
    #(ii)
    
    #value not found at first node
    #temp holds the value of node that will be deleted
    prev = temp          #change!!!
    temp = temp.next     #change!!!

    
    while temp != self.head: #change!!!
        if temp.val == val:
            break
        prev = temp
        temp = temp.next

    if temp == self.head: #not found
        return

    prev.next = temp.next #Just  lose the reference to delete node

Ring.remove = remove
```

```python
print(r)
r.remove(2)
print(r)
```
