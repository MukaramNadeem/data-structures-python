```python
a = 5
b = 4

if a > 2 and a < 10 and b > 0 and b < 10:
    print("All constraints satisfied")

#one method to check multiple if conditions
```

```python
#better way

a = 5
b = 4

constraints = { a > 2 , a < 10 , b > 0 , b < 10}
```

```python
if all(constraints):
    print("All constraints satisfied")
```

```python
if any(constraints):
    print("At least one constraint satisfied")
```

```python
X = {10,20,30,40,50}
```

```python

# ∀x ∈ X , x < 100


all([ x<100 for x in X]) #note its a list : iterable
```

```python
# ∃x ∈ X , x < 9
any([x<9 for x in X])
```

```python
def isDivisibleBy20(val):
    return val % 20 == 0
```

```python
all([isDivisibleBy20(x) for x in X])
```

```python
any ([isDivisibleBy20(x) for x in X])
```

```python
from collections import Counter
counter = Counter()

for word in ["red","green","blue","blue","red","blue"]:
    counter[word] +=1 
```

```python
counter
```

```python
counter["orange"]# no "key Error" : just returned zero
```

```python
counter.most_common() #prints dict in an ordered way
```

```python
counter.most_common()[1][0] #getting value from inside
```

```python
counter.most_common()[-1][0] #getting the last value : least common
```

```python
from collections import Counter
cnt = Counter()
sentence = """The "there exists" symbol is ∃, a backward capital 'E', 
                used in logic and mathematics to mean "there is at least one" or 
                "for some," asserting the existence of an element 
                satisfying a condition, like ∃x P(x)
                meaning "there exists an x such that P(x) is true".
                It's the existential quantifier, contrasted with ∀ 
                ("for all"), and signifies that a statement holds for
                one or more members of a set. """

for word in sentence.split():
    cnt[word] +=1
```
