```python
class Node:   #node has three things in Doubly
    def __init__(self,val = None ):     
        self.val  = val
        self.next = None    #this "next" holds the address of next node
        
        self.prev = None    #change !!! #this "prev" holds the address of previous node

        
class Doubly:
    def __init__(self):
        self.head = None
        
```

```python
def push(self,val): #push returns nothing
    new_node = Node(val) 
     #case (i)
    if self.head is None:
        self.head = new_node
        return

    
    #otherwise case (ii)
 
    last = self.head
    while last.next is not None:
        last = last.next

    
    last.next = new_node
    new_node.prev = last    #change !!!


Doubly.push = push
```

```python
def pop(self): #pop returns a value #Pop has no changes in Doubly
    #(i)
    if self.head is None:
        raise Exception ("No value in the list.")
    
    #(ii)
    if self.head.next is None:
        val = self.head.val
        self.head = None     #make the list empty
        return val
    
    #(iii) when there are more than one nodes. Use "two" pointers: make "prev" = temp and "temp" = next
    temp = self.head
    while temp.next is not None:
        prev = temp
        temp = temp.next
    #return the last value and make its next None using prev.next
    val = temp.val
    prev.next = None
    return val


Doubly.pop = pop
```

```python
def insert(self,index,val):
    new_node =  Node(val)

    #(i)
    if index == 0:
        new_node.next = self.head #do this first otherwise you will loss the head (whole Linked list). 
        
        
        if self.head is not None: #change!!! only this line(check) is new in case (i).
        
            self.head = new_node
        
            return

    
    #(ii)
    temp = self.head

    counter = 0
    while temp is not None and counter < index: #focus on "and" : Loop will run till the given index tion and the end of the List.First check "None" condition to deal with this condion insert(10000,1).
        prev = temp
        temp = temp.next
        counter +=1

        
    prev.next = new_node
    new_node.prev = prev #change!!!

    
    new_node.next = temp
    
    if temp is not None:     #change!!! 
        temp.prev = new_node #change!!!
    
    #(iii) It will be aumatically handled by case(ii) logic because of "None".
        
Doubly.insert = insert 
```

```python
def remove(self,val):
    
    temp = self.head

    #(i)
    if temp is not None:
        if temp.val == val:
            self.head = self.head.next #change!!!
            if self.head is not None:  #change!!!
                self.head.prev = None   
                return
    
    
    #(ii)
    while temp is not None:
        if temp.val == val:
            break #value found

        prev = temp
        temp = temp.next
    #after loop we reached the targeted position : temp is at target and prev is at previous node
    if temp is None: #not found after loop execution
        return

    
    prev.next = temp.next  #or prev.next = prev.next.next
    
    if temp.next is not None: #change!!!
        temp.next.prev = prev #change!!!


Doubly.remove = remove
```

```python
def printList(self):
    ret_str = "["
    
    temp = self.head
    while temp is not None:
        
        print(hex(id(temp.prev)),hex(id(temp)),hex(id(temp.next)))
            
        ret_str += str(temp.val) + ","
        temp = temp.next
    
    ret_str = ret_str.rstrip(', ')
    ret_str += "]"
    return ret_str
Doubly.__str__ = printList
```
