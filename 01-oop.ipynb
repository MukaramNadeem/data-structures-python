```python
class Point:
    def __init__(self, x=0,y=0,z = 0):       #constructor  
        self.x = x
        self.y = y
        self.z = z
        
    def __str__(self):                        #toString
        return "[" + str(self.x) + "," +  str(self.y) + "]"
```

```python
p1 = Point(3,5,6)
print(p1.x , p1.y ) 

p2 = Point()
print(p2.x,p2.y)

p3 = Point(4,3)
print(p3.y)
```

```python
print(p2) # test __str__
```

```python
class Shape:
    def __init__(self,points):
        self.points = points

    def __str__(self):
        ret = ""
        
        for i in self.points:
            ret+= str(i) + "-"
            
        return ret
```

```python
p1 = Point(5,5)
p2 = Point(10,5)
p3 = Point(5,10)
p = [p1,p2,p3]

sh = Shape(p)
```

```python
print(sh)
```

```python
def printPoints(self):
    for i in self.points:
        print(i)
    
Shape.runtime_method_to_class = printPoints
```

```python
sh.runtime_method_to_class()
```

```python
class Triangle (Shape):    #no extends
    pass
```

```python
t = Triangle(p)
```

```python
t.runtime_method_to_class()
```

```python
class Rectangle:
    def __init__(self,length,wedth):
        self.length = length
        self.wedth = wedth
        
    def area(self):
        return self.length * self.wedth
    
    def perimeter(self):
        return 2 * (self.length + self.wedth)

    def __str__(self):
        return "L: " + str(self.length) + " W: " + str(self.wedth)
    
        
```

```python
r = Rectangle(2,4)
print(r)
```

```python
class Square (Rectangle):
    def __init__(self,length):
        super().__init__(length, length)

    def __str__(self):
        return "Square: " + super().__str__()
```

```python
s = Square(4)
print(s)
s.area()
```
