```python
def fib1(n):
    if n<=1:
        return 1
    else:
        return fib1(n-2) + fib1(n-1)
```

```python
def fib2(n):
    a = 1
    b = 1
    for i in range(n):
        a , b = b , a+b
    
    return a
    
```

```python
print("fib1: ")
for i in range(15):
    print(fib1(i) , end =  ",  ")
print('\n')
print("fib2: ")
for i in range(15):
    print(fib2(i) ,  end = ",  ")
```

```python
import time

def compute_time(fun , limit):
    l = []
    for i in range (limit):
        start_time = int(round(time.time() * 1000 ))
        fun(i)
        end_time =   int(round(time.time() * 1000 ))
        time_taken = end_time - start_time
        
        l.append(time_taken)
        
    return l
```

```python
limit = 30
fib1_time  = compute_time(fib1 ,limit )
fib2_time  = compute_time(fib2 ,limit )
```

```python
def print_list(list):
    for i in range(limit):
        print(list[i] , end = " " )
```

```python
 print_list(fib1_time)
```

```python
print_list(fib2_time)
```
