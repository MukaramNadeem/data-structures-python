# Heap Data structure - MAX and MIN Heap





## Complete Binary Tree : 
#### Its all nodes have 2 children except the leafs : If any node has one child then it will be on left.

##### We can heapify a Complete binary Tree But here we will First Convert the Complete Binary Tree into a list(that will have left and right similar to that of a

##### real complete Binary Tree) then Heapify it: Its Efficient.  "OR Simply we just made the list a Complete Binary Tree (making the list  to have left and 

##### right indeces) and then just heapified it AND then sorted it as given below in our example." !!!

## Heap is formed by Complete Binary Tree:
#### It has two types MAX heap(greatest node at root) and MIN heap(Minimum node at the root) : It has nothing to do with the nodes present inside the tree.If root node is largest (MAX Heap) if smallest (Min Heap).
#### Heap is just to convert Complete Binary Tree into a Tree with its Root node largest (MAX Heap) or smallest (Min Heap) than all the members of the tree.

````python
def heapify (lst , n , root): #Its a max Heap logic. Its logic is simple, Heapy this Just one node and heapify the node changed or swaped recursively.We will use this function to heapify the whole tree later.
    #heapify function just check both the children of a node and then if any of the child is greater then it just replaces the value of that greater child node with itself and call the heapify function on that greater child node.
    """Heapify the root element of the lst which has n elements in Total"""
    #Root is the one index of the list.
    #Root here is the node going to be heapified. It could be other nodes(indeces) as we did in the loop later.

    
    #In this step we are actually making a list into a heap by identifing its left and right.
   
    
    l = 2*root +1                         #Method to find Left child of a node inside a list(list here is made up Complete Binary Tree or we can say represent it).
    r = 2*root +2                         #Method to find Right child of a node inside a list.

    #Now as we have got the left and right index of the  
    
    largest = root #to compare it with left and right we shored root into another variable.
    
    if l < n and lst[l] > lst[largest]: #root(current index) ka left agr list ka root sa bara ha to left ko largest variable ko upgrade kr do.
        largest = l
    
    
    if r < n and lst[r] > lst[largest]: #to make it  a MIN heap just chande this only sigh ( lst[r] ">" lst[largest] )  in both left and right checks as ```lst[r] < lst[largest]```
        largest = r

    #Now if vaue of largest variable got changed means there was at least one child greater than the root.
    #In this case largest will no longer be == root:
    if largest != root:
        
        lst[root] , lst[largest] = lst[largest] , lst[root] #Swap!!!

        heapify(lst,n,largest) #recursion to make the swaped node heapify. This recursion may continue to those lefts or rights that may not exist (and this will defenately happen) this we has a check `if r < n` and `if l < n` above
````

````python
def build_heap(lst):
    """Construct a heap from any list by repeatedly applying heapify from Bottom except leafs"""

    n = len(lst)
    #to keep it simople we will go reversed as it is easy.
    for i in reversed(range (n//2) ): # we do this because the other half of the list will always be a leaf : THE // RETURNS  a floor value not the float value.
        
        heapify(lst, n , i) #we are passing the index !!! Not value...
````

````python
heap = [100,5,3,2,8,15,6,102]
print(heap)
build_heap(heap)
print(heap)
````
